```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('15m_zams/15m_history.data', sep = ' ')
```

```python
df = df.iloc[:, 1:62]
```

```python
df
```

```python
df.info()
```

```python
fig = plt.figure()
plt.plot(df['star_age'], df['log_Teff'], label = 'PMS to ZAMS')
plt.xlabel('Star Age (years)')
plt.ylabel('log $T_{eff}$')
plt.legend()
plt.tight_layout()
```

```python
fig = plt.figure()
plt.plot(df['log_Teff'], df['log_L'], label = 'PMS to ZAMS')
plt.xlim(plt.xlim()[::-1])
plt.title('HR Diagram 15 $M_{sun}$')
plt.xlabel('log $T_{eff}$')
plt.ylabel('log L')
plt.legend()
plt.tight_layout()
```
