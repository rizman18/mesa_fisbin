# Profiles.data

```python
# load entire LOG directory information
l = mr.MesaLogDir('LOGS')
# grab the last profile
p = l.profile_data()

# this works even if you only have logRho and logT!
figure()
#loglog(p.Rho, p.T)
plot(p.zone, p.x_mass_fraction_H)
xlabel("Density")
ylabel("Temperature")
grid()
```
